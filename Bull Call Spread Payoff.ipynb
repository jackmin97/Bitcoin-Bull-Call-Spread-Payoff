# 1. Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
```

# 2. Call payoff

```python
def call_payoff(St, strike_price, premium):
    return np.where(St > strike_price, St - strike_price, 0) - premium
```

# 3. Define parameters

```python
# Bitcoin price
spot_price = 17100

# Long call
strike_price_long_call = 17400
premium_long_call = 300

# Short call
strike_price_short_call = 17860
premium_short_call = 285

# Bitcoin price range at expiration of the call
St = np.arange(0.95*spot_price, 1.1*spot_price, 1)
```
